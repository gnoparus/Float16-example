```python
FLOAT16_API_KEY = 'float16-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx'
FLOAT16_CUSTOM_URL = 'https://api.float16.cloud/v1/chat/completions'
```

```python
def init_system_task(prompt):
    output = []
    output.append({
        'role' : 'system',
        'content' : prompt
    })
    return output

def add_fewshot(data,example,label) :
    output = []
    output = data
    output.append({
        'role' : 'user',
        'content' : example
    })
    output.append({
        'role' : 'assistant',
        'content' : label
    })
    return output

def add_question(data,question) :
    output = []
    output = data
    output.append({
        'role' : 'user',
        'content' : question
    })
    return output
```

```python
import requests
import json
import os

prompt = ""
pair_data = [{
        'example' :"""
        นักเตะที่ทำผลงานได้ดีที่สุดที่ผ่านมา
        """,
        'label' : """
        {
          "season" : null,
          "bio" : null,
          "match" : null
        }
        """
        },
        {
        'example' :"""
        นักเตะต่างชาติที่ทำผลงานได้ดีที่สุด
        """,
        'label' : """
        {
          "season" : null,
          "bio" : foreign,
          "match" : null
        }
        """
        },
        {
        'example' :"""
        นักเตะญี่ปุ่นที่ทำผลงานได้ดีที่สุดในฤดูกาล 2023 ทั้งหมด
        """,
        'label' : """
        {
          "season" : 2023,
          "bio" : local,
          "match" : all
        }
        """
        },
        {
        'example' :"""
        นักเตะญี่ปุ่นที่ทำผลงานได้ดีที่สุดช่วง 10 นัดแรกในฤดูกาล 2022
        """,
        'label' : """
        {
          "season" : 2023,
          "bio" : local,
          "match" : 1-10
        }
        """
        },
    ]
def do_request(url, data):
    headers = {'content-type': 'application/json', 'Authorization': 'Bearer '+FLOAT16_API_KEY}
    response = requests.post(url, headers=headers, data=json.dumps(data))
    response_text = response.json()
    return response_text

message = init_system_task(''' ''')
for i in pair_data:
    message = add_fewshot(message, i['example'],i['label'])
message = add_question(message,"""
นักเตะที่ทำผลงาน 5 นัดแรกในฤดูกาล 2024 ทั้งหมด
""")
data = {
"messages" : message,
"model" : 'SeaLLM-7b-v2',
"max_tokens" : 1024,
"temperature" : 0.1
}
print(data)
res = do_request(FLOAT16_CUSTOM_URL, data)
print(res['choices'][0]['message']['content'])
```
