```python
FLOAT16_API_KEY = 'float16-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx'
FLOAT16_CUSTOM_URL = 'https://api.float16.cloud/v1/chat/completions'
```

```python
import requests
import json
import time

headers = {
    "Content-Type": "application/json",
}
messages = [{
    'role' : 'system',
    'content' : 'You are Thai people assistance.'
},{
    'role' : 'user',
    'content' : 'สวัสดี คุณรู้จักร้านอาหารในกรุงเทพมากน้อยแค่ไหน และคุณช่วยแนะนำนร้านนอาหารให้ผมได้รึเปล่า'
}]

data = {
    "messages" : messages,
    "model" : 'SeaLLM-7b', # Case-insensitive
    "max_tokens" : 1024,
    "temperature" : 0.1
}
```

```python
start = time.perf_counter()
e2e_start = time.time()
response = requests.post(FLOAT16_CUSTOM_URL, json=data, headers=headers,stream=True)

res = ''
if response.status_code == 200:
    for line in response.iter_lines():
        if line:
            data_str = line.decode('utf-8')
            json_data_str = data_str.split('data: ')[1]
            data_dict = json.loads(json_data_str)
            data_text = data_dict['text_output']
            new_text = data_text.replace("\n","")
            res += new_text
            print(f'{time.perf_counter()-start},{res},{new_text}')
else:
    print(f"Request failed with status code: {response.status_code}")
```
